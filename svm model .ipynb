```python
frames = [data1, data2, data3,data4]
data = pd.concat(frames)
data
```

```python
#drop the first column
data = data.drop('Time(s)', 1)
```

```python
#convert the sting values to float values
for i in range(len(data['250C'])):
    data['250C'].iloc[i]=float(data['250C'].iloc[i].replace(",", "."))
    data['300C'].iloc[i]=float(data['300C'].iloc[i].replace(",", "."))
    data['350C'].iloc[i]=float(data['350C'].iloc[i].replace(",", "."))
    data['200C'].iloc[i]=float(data['200C'].iloc[i].replace(",", "."))
```

```python
#final data
data
```

# SVM model 

```python
X = data.drop('class', axis=1)#select the feature
y = data['class']#select the label
X=preprocessing.normalize(X)#normalize the feature
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)#split the data into 80% training and 20% test 
```

```python
svclassifier = SVC(kernel='linear')#make the model
svclassifier.fit(X_train, y_train)#fit training data using the model
```

```python
y_pred = svclassifier.predict(X_test)#prediction
```

```python
#get the accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
